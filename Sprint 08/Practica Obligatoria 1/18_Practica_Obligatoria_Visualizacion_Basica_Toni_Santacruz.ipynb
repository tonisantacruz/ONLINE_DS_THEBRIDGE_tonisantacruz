### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)
2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)
3. El scatter de área de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
print("Columnas del archivo:", df.columns)

ciudades_top = df.nlargest(5, "population_total")
print(ciudades_top)
```

```python
plt.figure(figsize=(10, 6))  
plt.bar(ciudades_top["city"], ciudades_top["population_total"], color=["blue", "gray", "olive", "brown", "orange"])

for i, poblacion in enumerate(ciudades_top["population_total"]):
    plt.text(i, poblacion + 5000, str(poblacion), ha="center")
plt.title("Ciudades más pobladas de California")
plt.xlabel("Ciudades")
plt.ylabel("Población")
plt.show()
```

```python
plt.figure(figsize=(10, 6))  
plt.hist(df["area_total_km2"], bins=50, color="steelblue", edgecolor="black")  

plt.title("Distribución del área total ocupada por las ciudades", fontsize=18, fontname="Bradley Hand ITC")
plt.xlabel("Área total (km²)", fontsize=14)
plt.ylabel("Frecuencia", fontsize=14)

plt.show()
```

```python
plt.figure(figsize=(10, 6))  
plt.scatter(df["area_water_km2"], df["population_total"], color="dodgerblue", alpha=0.7, edgecolor="black")

plt.title("Relación entre el área de agua y la población de las ciudades", fontsize=16)
plt.xlabel("Área de agua (km²)", fontsize=14)
plt.ylabel("Población total", fontsize=14)
plt.grid(True, linestyle="--", alpha=0.5)

plt.show()
```

```python
plt.figure(figsize=(10, 6))  
plt.boxplot(df["elevation_m"], vert=False, patch_artist=True, boxprops=dict(facecolor="red", color="black"))


plt.title("Diagrama de caja de la Altitud de las ciudades de California", fontsize=16)
plt.xlabel("Altitud (m)", fontsize=14)
plt.yticks([1], ["Altitud"])  

plt.show()
```
