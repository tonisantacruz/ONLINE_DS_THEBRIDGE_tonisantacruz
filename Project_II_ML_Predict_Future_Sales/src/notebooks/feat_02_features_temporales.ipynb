```python
import pandas as pd
```

```python
df_matrix = pd.read_csv('../data/matriz_features_basicas.csv')
```

```python
def create_lag_feature(df, lag_months, col_name):
    for lag in lag_months:
        lag_col = f'{col_name}_lag_{lag}'
        df_lag = df[['date_block_num', 'shop_id', 'item_id', col_name]].copy()
        df_lag['date_block_num'] += lag
        df_lag.rename(columns={col_name: lag_col}, inplace=True)
        df = df.merge(df_lag, on=['date_block_num', 'shop_id', 'item_id'], how='left')
    return df
```

```python
#Crear lag de 1 y 3 meses para la variable objetivo (item_cnt_month)
df_matrix = create_lag_feature(df_matrix, [1, 3], 'item_cnt_month')

#Lag de media por producto
df_matrix = create_lag_feature(df_matrix, [1], 'item_avg_monthly_sales')

#Lag de media por tienda
df_matrix = create_lag_feature(df_matrix, [1], 'shop_avg_monthly_sales')
```

```python
df_matrix.fillna(0, inplace=True)
```
